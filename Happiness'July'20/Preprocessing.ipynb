## Data Cleaning
- 'Burundi','Jamaica','Somalia','Maldives','Trinidad and Tobago' have started the survey process recently so they had data for just 2019. Similarly, 'Congo (Kinshasa)','Malaysia','Comoros','Central African Republic','South Sudan','Swaziland','Bahrain' have recorded data from 2018 and 2019.
- Haiti,Hong Kong S.A.R. of China,Sri Lanka,Lesotho,Mozambique,Paraguay,Laos,North Cyprus,Iceland             ,Kuwait,Namibia,Czech Republic,Pakistan,Gambia,Israel,Russia,Iraq,Armenia had no data for 2016 but have still been included in the analysis.
- Decmocratic Quality and Delivery Quality were missing for the year of 2019 for every country so they were replaced by the countrywise-means for 2016-2018. 
- The handful of missing feature values still left were for countries that had missing data of that particular feature value for all 4 years. As a last resort these were filled up with feature wise averages.

```python
# Substituting Countrywise means for features missing for a given year 
df_2['Democratic Quality'] = df_2['Democratic Quality'].fillna(df_2.groupby('Country name')['Democratic Quality'].transform('mean'))
df_2['Delivery Quality'] = df_2['Delivery Quality'].fillna(df_2.groupby('Country name')['Delivery Quality'].transform('mean'))
df_2['GINI index (World Bank estimate), average 2000-2017, unbalanced panel'] = df_2['GINI index (World Bank estimate), average 2000-2017, unbalanced panel'].fillna(df_2.groupby('Country name')['GINI index (World Bank estimate), average 2000-2017, unbalanced panel'].transform('mean'))
df_2['Log GDP per capita'] = df_2['Log GDP per capita'].fillna(df_2.groupby('Country')['Log GDP per capita'].transform('mean'))
df_2['Social support'] = df_2['Social support'].fillna(df_2.groupby('Country name')['Social support'].transform('mean'))
df_2['Healthy life expectancy at birth'] = df_2['Healthy life expectancy at birth'].fillna(df_2.groupby('Country name')['Healthy life expectancy at birth'].transform('mean'))

df_2['Freedom to make life choices'] = df_2['Freedom to make life choices'].fillna(df_2.groupby('Country name')['Freedom to make life choices'].transform('mean'))
df_2['Healthy life expectancy at birth'] = df_2['Healthy life expectancy at birth'].fillna(df_2.groupby('Country name')['Healthy life expectancy at birth'].transform('mean'))
df_2['Healthy life expectancy at birth'] = df_2['Healthy life expectancy at birth'].fillna(df_2.groupby('Country name')['Healthy life expectancy at birth'].transform('mean'))
df_2['Healthy life expectancy at birth'] = df_2['Healthy life expectancy at birth'].fillna(df_2.groupby('Country name')['Healthy life expectancy at birth'].transform('mean'))
df_2['Healthy life expectancy at birth'] = df_2['Healthy life expectancy at birth'].fillna(df_2.groupby('Country name')['Healthy life expectancy at birth'].transform('mean'))


```
